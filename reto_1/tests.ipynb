```python
import pandas as pd

# Crear un DataFrame de ejemplo
data = {'Nombre': ['Juan', 'Ana', 'Luis'],
        'Edad': [23, 30, 22],
        'Ciudad': ['Madrid', 'Barcelona', 'Sevilla']}

df = pd.DataFrame(data)

# Convertir el DataFrame en una cadena formateada
formatted_str = df.to_string(index=False, header=True, justify='left')

# Crear la línea superior e inferior
column_widths = [max(len(str(val)) for val in df[col]) for col in df.columns]
border_line = '+' + '+'.join(['-' * (width + 2) for width in column_widths]) + '+'

# Crear las líneas de la tabla con bordes
table_lines = []
table_lines.append(border_line)  # Línea superior

# Agregar los encabezados con bordes
header_line = '| ' + ' | '.join([col.ljust(width) for col, width in zip(df.columns, column_widths)]) + ' |'
table_lines.append(header_line)

# Agregar la línea que separa el encabezado
table_lines.append(border_line)

# Agregar los valores de las filas con bordes
for _, row in df.iterrows():
    row_line = '| ' + ' | '.join([str(val).ljust(width) for val, width in zip(row, column_widths)]) + ' |'
    table_lines.append(row_line)

# Agregar la línea inferior
table_lines.append(border_line)

# Escribir la tabla en un archivo de texto
with open('tabla_con_lineas.txt', 'w') as f:
    f.write("\n".join(table_lines))
```

# Física:

1. Movimiento rectilíneo uniformemente acelerado (MRUA) a = 2d/t^2:

* Variables: Tiempo, distancia recorrida, aceleración.

* Descripción: Estudiar el comportamiento de un objeto que se mueve con aceleración constante. Se mide el tiempo y la distancia recorrida para calcular la aceleración.

2. Ley de Ohm V = IR:

* Variables: Voltaje, corriente eléctrica, resistencia.

* Descripción: En este experimento, se mide el voltaje y la corriente en un circuito resistivo, para calcular la resistencia según la ley de Ohm.

3. Caída libre de un objeto v = gt:

* Variables: Tiempo, altura desde la que cae el objeto, velocidad final.
* Descripción: Se deja caer un objeto desde una altura conocida y se mide el tiempo que tarda en llegar al suelo para estudiar su velocidad final.

# Química:

1. Reacción ácido-base:

* Variables: Concentración de ácido, volumen de ácido, pH.

* Descripción: Se mide el pH de una solución al agregar diferentes volúmenes de un ácido con concentración constante.

2. Descomposición de peróxido de hidrógeno :

* Variables: Concentración de peróxido de hidrógeno, temperatura, velocidad de descomposición.

* Descripción: Se observa la descomposición del peróxido de hidrógeno en presencia de un catalizador, variando la concentración y la temperatura.

3. Velocidad de disolución de un sólido en agua:

* Variables: Temperatura, tamaño del sólido, tiempo de disolución.

* Descripción: Se mide el tiempo que tarda un sólido en disolverse en agua a diferentes temperaturas y con partículas de distintos tamaños.

# Biología:

1. Efecto de la luz en la fotosíntesis:

* Variables: Intensidad de luz, concentración de dióxido de carbono, cantidad de oxígeno producido.

* Descripción: Se investiga cómo cambia la tasa de fotosíntesis de las plantas a diferentes intensidades de luz y concentraciones de dióxido de carbono.

2. Crecimiento de plantas bajo diferentes condiciones:

* Variables: Tipo de suelo, cantidad de agua, temperatura.

* Descripción: Se evalúa cómo afectan distintas condiciones de suelo, riego y temperatura al crecimiento de una planta.

3. Efecto de la temperatura sobre la actividad enzimática:

* Variables: Temperatura, concentración enzimática, velocidad de la reacción.

* Descripción: Se mide la velocidad de una reacción catalizada por una enzima a diferentes temperaturas para entender cómo influye la temperatura en la actividad enzimática.
Cada uno de estos experimentos tiene un máximo de tres variables, lo que los hace fáciles de manejar y controlar en el contexto de un laboratorio educativo.
